## Combining Happiness report dataset from 2015-2020

[Data Source](https://www.kaggle.com/mathurinache/world-happiness-report)

```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
```

```python
data_2015 = pd.read_csv("raw_data/2015.csv")
data_2016 = pd.read_csv("raw_data/2016.csv")
data_2017 = pd.read_csv("raw_data/2017.csv")
data_2018 = pd.read_csv("raw_data/2018.csv")
data_2019 = pd.read_csv("raw_data/2019.csv")
data_2020 = pd.read_csv("raw_data/2020.csv")
```

```python
data = data_2015[['Country', 'Region', 'Happiness Rank']].merge(data_2016[['Country', 'Region', 'Happiness Rank']], left_on=['Country','Region'], right_on = ['Country','Region'], how='inner')
```

```python
display(data)
```

```python
data.rename(columns={'Happiness Rank_x':'2015', 'Happiness Rank_y':'2016'}, inplace=True)
```

```python
data = data.merge(data_2017[['Country', 'Happiness.Rank']], left_on=['Country'], right_on = ['Country'], how='inner')
```

```python
display(data)
```

```python
data.rename(columns={'Happiness.Rank':'2017'}, inplace=True)
```

```python
data = data.merge(data_2018[['Country or region', 'Overall rank']], left_on=['Country'], right_on = ['Country or region'], how='inner')
```

```python
display(data)
```

```python
data.drop(columns=['Country or region'], inplace=True)
data.rename(columns={'Overall rank':'2018'}, inplace=True)
```

```python
display(data)
```

```python
data = data.merge(data_2019[['Country or region', 'Overall rank']], left_on=['Country'], right_on = ['Country or region'], how='inner')
```

```python
display(data)
```

```python
data.drop(columns=['Country or region'], inplace=True)
data.rename(columns={'Overall rank':'2019'}, inplace=True)
```

```python
display(data)
```

```python
data_2020['Overall rank'] = data_2020.index + 1
```

```python
data = data.merge(data_2020[['Country name', 'Overall rank']], left_on=['Country'], right_on = ['Country name'], how='inner')
```

```python
display(data)
```

```python
data.drop(columns=['Country name'], inplace=True)
data.rename(columns={'Overall rank':'2020'}, inplace=True)
```

```python
data
```

```python
data_combined_years = data.melt(id_vars=["Country", "Region"], 
        var_name="year", 
        value_name="rank")
```

```python
display(data_combined_years)
```

```python
data.to_csv("processed_data/combined_happiness_dataset.csv", index=False)
```

```python
data_combined_years.to_csv("processed_data/combined_happiness_dataset_joined_years.csv", index=False)
```
